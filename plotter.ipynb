```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "results.csv"
df = pd.read_csv(file_path)
```

```python

# Graficar con escala logartmica en el eje Y
plt.figure(figsize=(10, 6))
plt.bar(df['alg'], df['t_mean'], color='lightcoral')
plt.xlabel("Algoritmo")
plt.ylabel("Tiempo promedio (microsegundos) [escala logartmica]")
plt.yscale('log')
plt.title("Comparacin de tiempo promedio por versin del algoritmo (escala logartmica)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7, which='major')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.bar(df['alg'], df['memoria_usada(kb)'], color='mediumseagreen')
plt.xlabel("Algoritmo")
plt.ylabel("Memoria usada (kB)")
plt.yscale('log')
plt.title("Uso de memoria por versin del algoritmo")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
