```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "results.csv"
df = pd.read_csv(file_path)
```

```python

# Graficar con escala logarítmica en el eje Y
plt.figure(figsize=(10, 6))
plt.bar(df['alg'], df['t_mean'], color='lightcoral')
plt.xlabel("Algoritmo")
plt.ylabel("Tiempo promedio (microsegundos) [escala logarítmica]")
plt.yscale('log')
plt.title("Comparación de tiempo promedio por versión del algoritmo (escala logarítmica)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7, which='major')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.bar(df['alg'], df['memoria_usada(kb)'], color='mediumseagreen')
plt.xlabel("Algoritmo")
plt.ylabel("Memoria usada (kB)")
plt.yscale('log')
plt.title("Uso de memoria por versión del algoritmo")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for alg in df['alg'].unique():
    subset = df[df['alg'] == alg].copy()
    subset["comp"] = subset["txt1"] + " vs " + subset["txt22"]
    plt.plot(subset['comp'], subset['t_mean'], marker='o', label=alg)

plt.xlabel("Comparación de textos")
plt.ylabel("Tiempo promedio (nanosegundos)")
plt.yscale('log')
plt.title("Tiempo promedio por algoritmo")
plt.legend(title="Algoritmo")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("TiempoPorAlgoritmo")
plt.show()
```
